# Conditional Statements in Python

Conditional statements in Python allow you to execute certain blocks of code based on whether a condition is true or false.

## The `if` statement

The `if` statement is used to test a specific condition. If the condition is true, a block of code (the if-block) is executed.

```python
age = 20
if age >= 18:
    print("You are allowed to vote")
```

## The `if-elif-else` statement

The `if-elif-else` statement provides a way to check multiple conditions.

```python
age = 18

if age < 13:
    print("You are a child")
elif age < 18:
    print("You are a teenager")
else:
    print("You are an adult")
```

## Nested `if-else`

You can have `if` statements inside other `if` statements, this is called nesting.

```python
num = int(input("Enter a number: "))
if num > 0:
    print("It is positive")
    if num % 2 == 0:
        print("It is Even")
    else:
        print("It is odd")
elif num == 0:
    print("It is zero")
else:
    print("It is negative")
```

## Finding the largest among three numbers

```python
a, b, c = 2, 4, 6
largest = a if a > b else b
largest = largest if largest > c else c

print("largest is:", largest)
```

## Simple calculator

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
operator = input("Enter the operation you want to perform (+, -, *, /): ")

if operator == '+':
    print("Sum of the number is", num1 + num2)
elif operator == '-':
    print("Difference between the number is", num1 - num2)
elif operator == '*':
    print("Multiplication of the number is", num1 * num2)
elif operator == '/':
    if num2 != 0:
        print("Division of the number is", num1 / num2)
    else:
        print("Error: Division by zero is not allowed.")
else:
    print("Unknown Operator")
```
