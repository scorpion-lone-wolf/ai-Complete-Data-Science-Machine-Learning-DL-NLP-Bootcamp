```python
# largest number among 3 numbers
a,b,c=2,4,6
largest = a
largest = a if a>b else b
largest= largest if largest>c else c

print("largest is:",largest)
```

```python
# if , elif , else
age=18

if age<13:
    print("You are a child")
elif age<18:
    print("You are a teenager")
else:
    print("You are an adult")
```

```python
# nested if else
num=int(input("Enter a number"))
if num>0:
    print("It is positive")
    if num%2==0:
        print("It is Even")
    else:
        print("It is odd")
else:
    print("It is zero or negative")
```

```python
## Simple calculator
num1=int(input("Enter first number"))
num2=int(input("Enter second number"))
operator=input("Enter the operation you want to perform. (+,-,*,/)")

if operator=='+':
    print("Sum of the number is", num1+num2)
elif operator=='-':
     print("Difference between the number is", num1-num2)
elif operator=='*':
      print("Multiplication of the number is", num1*num2)
elif operator=='/':
     print("Division of the number is", num1/num2)
else:
     print("Unknown Operator")


```
