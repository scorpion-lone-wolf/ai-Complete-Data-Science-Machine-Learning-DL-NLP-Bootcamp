# Loops in Python

## The `for` Loop

A `for` loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).

### The `range()` function

````python
# The range() function returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and stops before a specified number.
range(5)
````

**Syntax of `range()`
```python
# range(start, end, increment)
```
Here `end` is exclusive.

````python
for i in range(5):
    print(i)
````

````python
for i in range(1, 10, 2):
    print(i)
````

Loop from the back side:

````python
for i in range(6, 0, -2):
    print(i)
````

### Looping in string

````python
name = "rahul singh"
for ch in name:
    print(ch)
````

## The `while` loop

With the `while` loop we can execute a set of statements as long as a condition is true.

````python
count = 5
while count > 0:
    print(count)
    count -= 1
````

## `break`, `continue` & `pass`

### Example for `break`

````python
for i in range(5):
    if i == 2:
        break
    print(i)
````

### Example for `continue`

````python
for i in range(5):
    if i == 2:
        continue
    print(i)
````

### Example for `pass`

````python
for i in range(5):
    if i == 2:
        pass
    print(i)
````

## Nested loop

````python
for i in range(2):
    for j in range(2):
        print(f"i is -> {i}, j is -> {j}")
````

## Prime number

````python
for i in range(2, 25):
    if i > 1:
        for j in range(2, i):
            if i % j == 0:
                break
        else:
            print(i)
````

## Sum on n natural number

````python
n = 10
sum_of_n = 0
for i in range(n + 1):
    sum_of_n += i
print("Sum of first ".n, " natural number is :", sum_of_n)
````
