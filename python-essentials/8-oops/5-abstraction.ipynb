#### Abstraction
- Abstraction is the concept of hiding the complex implementation details and showing only the necessaey features of an object.
- This helps in reducing programming complexity and effort.

```python
from abc import ABC, abstractmethod
# Abstract base class
class Vehicle(ABC):
    # This is an abstract method that must be implemented by any subclass
    @abstractmethod
    def drive(self):
        pass
    # This is a concrete method that can be used by subclasses
    def info(self):
        return "This is a vehicle."
```

```python
class Car(Vehicle):
    def drive(self):
        return "The car is driving."

class Motorcycle(Vehicle):
    def drive(self):
        return "The motorcycle is driving."
# Create instances of Car and Motorcycle
car = Car()
motorcycle = Motorcycle()
# Call the drive method and info method for both instances
print(car.drive())  # Output: The car is driving.
print(car.info())   # Output: This is a vehicle.
print("===============")
print(motorcycle.drive())  # Output: The motorcycle is driving.
print(motorcycle.info())   # Output: This is a vehicle.
```
