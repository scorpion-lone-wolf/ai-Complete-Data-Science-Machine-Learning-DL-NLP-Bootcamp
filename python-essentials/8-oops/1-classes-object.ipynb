A class is a blueprint for creating objects.

It defines a set of attributes and methods that the created objects will have.

In Python, you can define a class using the `class` keyword followed by the class name and a colon. Here's an example of a simple class definition:

```python
class Car:
    # constructor method to initialize the attributes of the class
    def __init__(self, name:str="unknown", model:str="unknown", year:int=0):
        self.name=name
        self.model=model
        self.year=year
        print("Car object created!")

    # instance method to display the details of the car
    def display_details(self):
        print(f"Car Name: {self.name}")
        print(f"Car Model: {self.model}")
        print(f"Car Year: {self.year}")

    def __internal_method(self):
        print("This is an internal method, not accessible outside the class.")
# Creating an object (instance) of the Car class
my_car1 = Car()
my_car2 = Car()
my_car3 = Car("Toyota", "Camry", 2020)
# * You can still access the internal method using name mangling, but it's not recommended as it breaks encapsulation.
my_car3._Car__internal_method()
my_car3.display_details()
```

```python
### Modeling a bank account using a class
class BankAccount:
    def __init__(self,account_holder:str="unknown", balance:float=0.0):
        self.account_holder=account_holder
        self.balance=balance
        print("BankAccount object created!")

    def deposit(self, amount:float):
        if amount < 0 :
            print("Deposit amount must be positive.")
            return
        self.balance += amount
        # showing only 2 decimal places for better readability
        print(f"Deposited ${amount:.2f}. New balance: ${self.balance:.2f}")

    def withdraw(self, amount:float):
        if amount == 0:
            print("Withdrawl amount must be greater than zero.")
            return
        if amount < 0 :
            print("withdrawl amount must be positive")
            return
        if amount > self.balance:
            print("Insufficient funds.")
            return
        self.balance -= amount
        print(f"Withdrew ${amount:.2f}. New balance: ${self.balance:.2f}")
    def display_balance(self):
        print(f"Account Holder: {self.account_holder}")
        print(f"Current Balance: ${self.balance:.2f}")
```

```python
rahul = BankAccount("Rahul", 1000.0)
rahul.display_balance()
rahul.deposit(500.0)
rahul.display_balance()
rahul.withdraw(200.0)
rahul.display_balance()
```
