### Inheritance
Inheritance is a fundamental concept in object-oriented programming (OOP) that allows a new class (called a child or subclass)
to inherit attributes and methods from an existing class (called a parent or superclass).
This promotes code reusability and establishes a natural hierarchical relationship between classes.

```python
# Single Inheritance Example
class BaseClass:
    def __init__(self,name:str):
        self.name = name
    def display_name(self):
        print(f"Name: {self.name}")

class DerivedClass(BaseClass):
    def __init__(self,name:str,age:int):
        super().__init__(name)  # Call the constructor of the base class
        self.age = age
    def display_age(self):
        print(f"Age: {self.age}")
    def display_info(self):
        self.display_name()  # Call method from base class
        self.display_age()   # Call method from derived class

```

```python
rahul = DerivedClass("Rahul", 28)
rahul.display_info()
```

```python
# Multiple Inheritance Example
class Parent1:
    def __init__(self, name:str):
        self.name = name
    def display_name(self):
        print(f"Name: {self.name}")
    def display_greeting(self):
        print(f"Hello, {self.name}!")
class Parent2:
    def __init__(self, age:int):
        self.age = age
    def display_age(self):
        print(f"Age: {self.age}")
    def display_greeting(self):
        print(f"Hi, you are {self.age} years old!")

class Child( Parent2,Parent1):
    def __init__(self, name:str, age:int):
        Parent1.__init__(self, name)  # Initialize Parent1
        Parent2.__init__(self, age)   # Initialize Parent2

    def display_info(self):
        self.display_name()  # Call method from Parent1
        self.display_age()   # Call method from Parent2
        self.display_greeting()  # This will call the method from Parent2 due to method resolution order (MRO)

    # def display_greeting(self):
    #     print(f"Hello, {self.name}! You are {self.age} years old!")

child = Child("Alice", 30)
child.display_info()
```

#### If Child has same method as Parent1 or Parent2, then the method in Child will override the method in Parent1 or Parent2. This is known as method overriding.
#### If Child Inherits same method from both Parent1 and Parent2, then the method in Parent1 will be called. This is known as method resolution order (MRO). i.e first parent class it inherits.
