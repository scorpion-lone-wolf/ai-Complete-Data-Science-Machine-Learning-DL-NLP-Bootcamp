```python
# ============================================================
# FILE & DIRECTORY OPERATIONS USING os AND pathlib
# ============================================================

import os
from pathlib import Path


# ------------------------------------------------------------
# 1. CREATE A DIRECTORY
# ------------------------------------------------------------

dir_name = "my_directory"

# Check if directory already exists to avoid error
if not os.path.exists(dir_name):
    os.mkdir(dir_name)
    print(f"Directory '{dir_name}' created.")
else:
    print(f"Directory '{dir_name}' already exists.")
```

```python
# ------------------------------------------------------------
# 2. LIST ALL FILES & DIRECTORIES
# ------------------------------------------------------------

print("\nContents of current directory:")
for item in os.listdir():
    print(item)
```

```python
# ------------------------------------------------------------
# 3. JOINING PATHS PROPERLY
# ------------------------------------------------------------

# Define path segments
segments = ["my_directory", "file.txt"]

# os.path.join ensures cross-platform compatibility
# *segments unpacks the list into separate arguments
file_path = os.path.join(os.getcwd(), *segments)

print("\nJoined File Path:")
print(file_path)
```

```python
# ------------------------------------------------------------
# 4. CREATE A FILE USING pathlib
# ------------------------------------------------------------

file_path = Path("my_directory/newfile.txt")

# Check if file exists
if file_path.exists():
    print("\nFile already exists.")
else:
    print("\nFile does not exist, creating file...")

    # Create an empty file
    file_path.touch()

    print("File created successfully.")
```
