# Python Standard Libraries

Short notes on a few useful built-in Python modules.

- Standard libraries come with Python.
- We import them when we need specific functionality.
- These examples cover arrays, random values, file handling, JSON, CSV, dates, and time.

## 1. `array`

The `array` module stores elements of the same data type in a compact form.

```python
import array

# Array of characters
arr_char = array.array('u', ['a', 'b', 'c'])

# Array of integers
arr_int = array.array('i', [1, 2, 3])

# Array of floats
arr_float = array.array('f', [1.1, 2.2, 3.3])

# Array of doubles
arr_double = array.array('d', [1.11, 2.22, 3.33])

# `b` stores signed integers, so booleans become 1 and 0
arr_bool_like = array.array('b', [True, False, True])

print('Characters:', arr_char)
print('Integers:', arr_int)
print('Floats:', arr_float)
print('Doubles:', arr_double)
print('Boolean-like values:', arr_bool_like)
```

## 2. `random`

Useful for generating random numbers or picking random values.

```python
import random

print('Random integer from 1 to 10:', random.randint(1, 10))
print('Random choice from list:', random.choice(['a', 'b', 'c']))
```

## 3. `os`

The `os` module helps us work with the operating system, such as folders and files.

```python
import os

folder_name = 'new_dir'

print('Current working directory:', os.getcwd())

# Create the directory only if it does not already exist
if not os.path.exists(folder_name):
    os.mkdir(folder_name)
    print(f"Created folder: {folder_name}")
else:
    print(f"Folder already exists: {folder_name}")
```

```python
# Remove the directory only if it exists
if os.path.exists(folder_name):
    os.rmdir(folder_name)
    print(f"Removed folder: {folder_name}")
else:
    print(f"Folder not found: {folder_name}")
```

## 4. `shutil`

The `shutil` module is used for higher-level file operations like copying files.

```python
import shutil

source_file = 'source.txt'
copied_file = 'copy.txt'

shutil.copy(source_file, copied_file)
print(f'Copied {source_file} to {copied_file}')
```

```python
# Clean up the copied file after the example
if os.path.exists(copied_file):
    os.remove(copied_file)
    print(f'Removed file: {copied_file}')
else:
    print(f'File not found: {copied_file}')
```

## 5. `json`

`json` is used to convert Python objects to JSON strings and back again.

```python
import json

data = {'name': 'John', 'age': 30, 'city': 'New York'}

# Convert Python dictionary to JSON string
json_data = json.dumps(data)
print('JSON data:', json_data)
print('Type after dumps:', type(json_data))

# Convert JSON string back to Python dictionary
python_data = json.loads(json_data)
print('Python data:', python_data)
print('Type after loads:', type(python_data))
```

## 6. `csv`

The `csv` module helps write and read comma-separated values files.

```python
import csv

with open('data.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['name', 'age', 'city'])
    writer.writerow(['John', 30, 'New York'])

print('CSV file written successfully.')
```

```python
with open('data.csv', mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

## 7. `datetime`

Use `datetime` when you want to work with dates and times.

```python
import datetime

now = datetime.datetime.now()
print('Current date and time:', now)

one_day_before = now - datetime.timedelta(days=1)
print('One day before:', one_day_before)
```

## 8. `time`

The `time` module is often used for delays and basic timing tasks.

```python
import time

print('Sleeping for 1 second...')
time.sleep(1)
print('Done sleeping!')
```

## Quick summary

- `array` stores fixed-type values.
- `random` generates random values.
- `os` and `shutil` help with file and folder operations.
- `json` and `csv` help with data storage and exchange.
- `datetime` and `time` help us work with time-related operations.
