### Generator
- A Generator is a special type of function that produces an iterable sequence of values on demand using the `yield` keyword instead of `return`.
- Unlike regular functions that executes entirely and store all result in memory, Generator pause execution at each yield statement, preserving their internal state

```python
def even_generator(start:int, end:int):
    for num in range(start,end+1):
        if num % 2 == 0:
            yield num
        num+=1
```

```python
my_generator = even_generator(1,10)
print(my_generator)
try:
    while True:
        print(next(my_generator))
except StopIteration:
    print("Generator exhausted")

```

```python
# Reading Large Files with Generators
def read_large_file(file_path):
    with open(file_path,'r') as file:
        for line in file:
            yield line.strip()  # Yielding one line at a time

# Example usage
file_path='example.txt'
for line in read_large_file(file_path):
    print(line)
```
