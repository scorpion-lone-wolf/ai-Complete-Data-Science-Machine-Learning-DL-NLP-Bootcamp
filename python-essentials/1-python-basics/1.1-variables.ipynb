### Variable Basics

```python
## Declaring and Assigning variables
age=28
height=5.4
name='rahul'
is_student=False

print("Name is:", name);
print("Age is:", age);
print("Height is:", height);
print("Is Student:", is_student);
```

```python
## Naming Convention
# Variable should be descriptive
# They must start with a letter or an '_' and contains letter , number and underscore
# Variable names are case sensitive

# Valid variable name
first_name='Rahul'
last_name='Singh'
```

```python
## Invalid number names
# 2name='rahul'
# first-name='rahul'
# @name='rahul'
```

### Variable Types

```python
## Understanding variables type
# Python is a dynamically typed language
# type of a variable is determined at runtime

age=28 # int
height=5.34 # float
name='rahul' # str
is_student=True # bool
```

```python
## Type Checking and Conversion

# Type checking
print(type(name)) # <class 'str'>
print(type(age)) # <class 'int'>
print(type(height)) # <class 'float'>
```

```python
# Type Conversion
# float to int
print(int(1.23))
# bool to int
print(int(False))
# str to int
print(int('23'))

# int to float
print(float(1))
```

### Dynamic Typing

```python
## Python allows the type of a variable to change as the program execute
var=10
print(var,type(var))

var='rahul'
print(var,type(var))
```

```python
num1 = int(input("Enter the first number"))
num2 = int(input("Enter the second number"))

#  doing some calculation
# addition
# subtraction
# multiplication
# exponential
sum = num1+num2
diff = num1-num2
mul = num1*num2
exp = num1**num2
print("Sum :", sum)
print("Diff :", diff)
print("Mul :", mul)
print("Exp :", exp)
```
