#### ⭐️ Arithmetic Operators

```python
a=10
b=5

add_result=a+b          # addition
sub_result=a-b          # subtraction
mult_result=a*b         # multiplication
div_result=a/b          # division - returns float
floor_div_result=a//b   # floor division - only take floor value
modulus_result=a%b      # returns the reminder
exponent_result=a**b    # a^b
xor_result=a^b
'''
a = 10 -> 1010
b = 5  -> 0101
 --------------
=         1111 in base 2 = 15 in base 10
'''

print(add_result)
print(sub_result)
print(mult_result)
print(div_result)
print(floor_div_result)
print(modulus_result)
print(exponent_result)
print(xor_result)
```

#### ⭐️ Logical Operator

```python
# and operator
print(0 and 0)
print(0 and 1)
print(1 and 0)
print(1 and 1)
```

```python
# or operator
print(0 or 0)
print(0 or 1)
print(1 or 0)
print(1 or 1)
```

```python
# not operator (negation)
print(not True) # False
print(not False) # True
```
