# Dictionaries in Python

Dictionaries are Python's implementation of a hash table, an unordered collection of items. They are mutable and store data in **key-value pairs**.

Key properties:
- **Keys** must be unique and immutable (e.g., strings, numbers, or tuples).
- **Values** can be of any data type and can be duplicated.
- Dictionaries are great for retrieving data quickly when you know the key.

## Creating Dictionaries

### Using Curly Braces `{}`

```python
my_dict = {'name': 'John', 'age': 30, 'city': 'New York'}
print(my_dict)
print(type(my_dict))
```

### Using the `dict()` Constructor

```python
my_dict2 = dict(name='Alice', age=25, city='Los Angeles')
print(my_dict2)
```

## Accessing Values

### Using Square Brackets `[]`
This is the most common way, but it will raise a `KeyError` if the key does not exist.

```python
print(my_dict['name'])
```

### Using the `.get()` Method
This method is safer as it returns `None` (or a specified default value) if the key is not found, avoiding an error.

```python
# Returns the value for 'age'
print(my_dict.get('age'))

# Returns the default value 'USA' because 'country' key does not exist
print(my_dict.get('country', 'USA'))
```

## Modifying Dictionaries
Dictionaries are mutable, so you can change them after they have been created.

```python
print("Original dictionary:", my_dict)

# Changing the value of an existing key
my_dict['age'] = 31
print("After changing age:", my_dict)

# Adding a new key-value pair
my_dict['country'] = 'USA'
print("After adding country:", my_dict)
```

## Dictionary Methods

```python
print("Keys:", my_dict.keys())      # Returns a view of the dictionary's keys
print("Values:", my_dict.values())  # Returns a view of the dictionary's values
print("Items:", my_dict.items())    # Returns a view of the dictionary's key-value tuple pairs
```

## Iterating Over Dictionaries

```python
# The .items() method is great for iterating because it gives you the key and value at the same time.
for key, value in my_dict.items():
    print(f"{key}: {value}")
```

## References vs. Copies
This is a very important concept. Understanding the difference between assigning, shallow copying, and deep copying a dictionary will save you from many headaches!

### 1. Assignment (Creating a Reference)
When you use the `=` operator, you are **not** creating a new dictionary. You are creating a second name (a reference) that points to the exact same dictionary in memory. Any changes made through one variable will affect the other.

```python
original_dict = {'name': 'John', 'age': 30}
referenced_dict = original_dict # This is just a reference, not a copy

print("Original Dictionary:", original_dict)
print("Referenced Dictionary:", referenced_dict)

# Now, change the referenced dictionary
referenced_dict['age'] = 99

print("\n--- After changing the reference ---")
# Notice the original dictionary ALSO changed
print("Original Dictionary:", original_dict)
print("Referenced Dictionary:", referenced_dict)
```

### 2. Shallow Copy (.copy())

**What it is:** A shallow copy creates a new dictionary, so the main dictionary is a separate object. 

**The Catch:** If the dictionary contains complex items (like a list or another dictionary), it does **not** create new versions of those inner items. It just copies the *reference* (the memory address) to them. 

**Analogy:** Imagine you photocopy a folder. You get a new folder, but if a page inside has a sticky note saying "See the shared binder on the shelf", your photocopy will also have a sticky note pointing to that *same single binder*. If someone changes the contents of the binder, the change is reflected for both the original folder and your new photocopy.

```python
original_dict_nested = {'name': 'Alice', 'grades': [90, 85]}
shallow_copy = original_dict_nested.copy()

print("Original:", original_dict_nested)
print("Shallow Copy:", shallow_copy)

# Let's change the list inside the shallow copy
shallow_copy['grades'].append(100)

print("\n--- After changing a nested item ---")
# The change appears in BOTH because they both reference the same list!
print("Original:", original_dict_nested)
print("Shallow Copy:", shallow_copy)
```

### 3. Deep Copy (copy.deepcopy())

**What it is:** A deep copy creates a new dictionary AND *recursively* creates new copies of all objects inside it, including any nested lists or dictionaries.

**The Result:** The new dictionary is a completely independent clone of the original. Nothing you do to the copy will ever affect the original, and vice-versa.

**Analogy:** Using the same folder example, a deep copy is like photocopying the folder, and then for every sticky note that says "See the shared binder on the shelf", you go to the shelf, photocopy the *entire binder*, and update the sticky note in your new folder to point to your new, separate binder.

```python
import copy

original_dict_nested = {'name': 'Bob', 'grades': [70, 80]}
deep_copy = copy.deepcopy(original_dict_nested)

print("Original:", original_dict_nested)
print("Deep Copy:", deep_copy)

# Modify the list inside the deep copy
deep_copy['grades'].append(95)

print("--- After changing a nested item ---")
# The change ONLY appears in the deep copy
print("Original:", original_dict_nested)
print("Deep Copy:", deep_copy)
```

## Nested Dictionaries
You can have dictionaries inside of other dictionaries. This is useful for storing complex, structured data.

```python
students = {
    'student1': {'name': 'Alice', 'age': 20, 'grade': 'A'},
    'student2': {'name': 'Bob', 'age': 22, 'grade': 'B'},
    'student3': {'name': 'Charlie', 'age': 21, 'grade': 'A'}
}

# Accessing data in a nested dictionary
print(students['student1']['name'])
print(students['student2']['grade'])
```

## Dictionary Comprehension
Similar to list comprehensions, dictionary comprehensions provide a concise way to create dictionaries.

```python
# Create a dictionary of numbers and their squares
square_dict = {x: x**2 for x in range(1, 6)}
print(square_dict)
```
