# Pandas Notes

These notes are organized to help you revise Pandas quickly during feature engineering and day-to-day data analysis.


## What is Pandas?

Pandas is a Python library built on top of NumPy for working with structured data.

It is mainly used for:
- reading datasets
- cleaning and transforming data
- selecting rows and columns
- summarizing and analyzing tabular data

### Two core data structures
- `Series`: one-dimensional labeled data
- `DataFrame`: two-dimensional table with rows and columns


## Setup

We start by importing Pandas. The alias `pd` is the standard convention used in almost every Pandas project.


```python
import pandas as pd

# Optional: make notebook tables slightly easier to read.
pd.set_option("display.max_columns", None)
```

## 1. Series

A `Series` is a one-dimensional labeled array.

You can think of it like a single column in a spreadsheet, where each value also has an index.


```python
# Creating a Series from a list.
# Pandas automatically assigns a default numeric index starting from 0.
numbers = [1, 2, 3, 4, 5]
series_from_list = pd.Series(numbers)

print("Series from list:\n")
display(series_from_list)
```

```python
# Creating a Series from a dictionary.
# Dictionary keys become index labels automatically.
marks_dict = {"Math": 85, "Science": 90, "English": 88}
series_from_dict = pd.Series(marks_dict)

print("Series from dictionary:\n")
display(series_from_dict)
```

```python
# Creating a Series with a custom index.
# This is useful when labels are more meaningful than numeric positions.
scores = [10, 20, 30]
labels = ["A", "B", "C"]
custom_series = pd.Series(data=scores, index=labels)

print("Series with custom index:\n")
display(custom_series)
```

## 2. DataFrame

A `DataFrame` is a two-dimensional tabular structure.

It is the most commonly used Pandas object because real datasets usually come in table form.


```python
# Creating a DataFrame from a dictionary of lists.
# Each key becomes a column.
people_data = {
    "Name": ["Rahul", "Pursotam", "Ram"],
    "Age": [24, 26, 28],
    "City": ["Kolkata", "New York", "Florida"]
}

people_df = pd.DataFrame(people_data)

print("DataFrame from dictionary of lists:\n")
display(people_df)
print("Type:", type(people_df))
```

```python
# Creating a DataFrame from a list of dictionaries.
# Here, each dictionary represents one row.
people_rows = [
    {"Name": "Rahul", "Age": 27, "City": "Kolkata"},
    {"Name": "Pursotam", "Age": 28, "City": "Bangalore"},
    {"Name": "Kane", "Age": 27, "City": "Florida"}
]

people_df = pd.DataFrame(people_rows, index=["First", "Second", "Third"])
display(people_df)
```

## 3. Reading a CSV File

One of the most common Pandas tasks is loading data from a CSV file.

In this notebook, we use `sales_data.csv` from the same folder.


```python
# Read the CSV file into a DataFrame.
sales_df = pd.read_csv("sales_data.csv")

# head() shows the first 5 rows by default.
sales_df.head()
```

```python
# Quick structural inspection of the dataset.
print("Column names:", sales_df.columns.tolist())
print("\nData types:\n", sales_df.dtypes)
print("\nShape of the DataFrame:", sales_df.shape)
```

```python
# Statistical summary for numeric columns.
# Useful for a quick understanding of count, mean, std, min, max, and quartiles.
sales_df.describe()
```

## 4. Selecting Data

Pandas makes it easy to select columns, rows, and even specific values.

This is one of the most important skills for data cleaning and feature engineering.


```python
# Selecting one column returns a Series.
display(people_df["Name"])
print("Type of single column:", type(people_df["Name"]))

# Selecting multiple columns returns a DataFrame.
display(people_df[["Name", "City"]])
```

```python
# loc[] is label-based indexing.
print("Row with label 'Second':")
display(people_df.loc["Second"])

# iloc[] is position-based indexing.
print("\nRow at position 0:")
display(people_df.iloc[0])
```

```python
# at[] accesses a single value using row label and column label.
print("Value using at[]:", people_df.at["First", "Name"])

# iat[] accesses a single value using numeric row and column positions.
print("Value using iat[]:", people_df.iat[0, 0])
```

## 5. Data Manipulation

Once the data is loaded, we often need to create columns, remove columns, or update existing values.

Below are a few basic operations you will use frequently.


```python
# Work on a copy so the original example stays unchanged.
people_modified_df = people_df.copy()
people_modified_df
```

```python
# Add a new column.
people_modified_df["Salary"] = [2000, 4000, 5000]
people_modified_df
```

```python
# Remove a column.
# axis=1 means we are dropping a column, not a row.
people_modified_df = people_modified_df.drop("Salary", axis=1)
people_modified_df
```

```python
# Update an existing column.
# Keep the column name exactly the same: 'Age', not 'age'.
people_modified_df["Age"] = people_modified_df["Age"] + 1
people_modified_df
```

## Quick Revision Summary

- Use `pd.Series()` to create a Series.
- Use `pd.DataFrame()` to create a DataFrame.
- Use `pd.read_csv()` to load CSV data.
- Use `head()`, `dtypes`, `shape`, and `describe()` for quick inspection.
- Use `[]`, `loc[]`, `iloc[]`, `at[]`, and `iat[]` to access data.
- Use assignment and `drop()` for basic data manipulation.
