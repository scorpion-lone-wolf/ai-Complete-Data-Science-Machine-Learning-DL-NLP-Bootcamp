```python
# Import necessary libraries
import pandas as pd
import numpy as np
```

## 1. Loading & Exploring Data

In this section, we'll load a CSV file and examine its structure to understand the data we're working with.

```python
# Load the CSV file into a Pandas DataFrame
df = pd.read_csv("data.csv")

# Display the first 5 rows to get an overview of the data
print("First 5 rows of the dataset:")
display(df.head())

print("\nLast 5 rows of the dataset:")
# Display the last 5 rows to check the end of the data
display(df.tail())
```

```python
# Check the data types of all columns in the dataset
# This shows whether columns are numeric, string, datetime, etc.
print("Data types of each column:")
print(df.dtypes)
```

### Understanding Data Types

It's important to know the data types of each column in your dataset. This helps identify potential issues and determines how to process the data correctly.

```python
# Check if there are any missing values in each column
# Returns True if the column has at least one missing value, False otherwise
print("Columns with missing values:")
print(df.isnull().any(axis=0))
```

## 2. Inspecting Missing Values

Missing values (NaN - Not a Number) can affect our analysis. It's crucial to identify and handle them appropriately.

```python
# Count the number of missing values in each column
# This gives us a summary of how many NaN values exist in each column
print("Count of missing values per column:")
print(df.isnull().sum())
```

```python
### Strategy 1: Fill Missing Values with a Constant (0)
# Create a new DataFrame with missing values filled with 0
df_filled = df.fillna(0)
print("First few rows after filling with 0:")
display(df_filled.head())
```

## 3. Data Cleaning - Handling Missing Values

There are several strategies to handle missing values:
- **Fill with a constant value** - Replace NaN with 0, -1, or other constants
- **Fill with statistical value** - Use mean, median, or mode to fill missing values
- **Forward/Backward fill** - Propagate values from adjacent rows
- **Drop rows/columns** - Remove rows or columns with missing values

```python
### Strategy 2: Fill Missing Values with Column Mean
# Calculate the mean of the 'Value' column
# Use this mean to fill any missing values in that column
# This preserves the distribution of the data better than using a constant
df['Value_fillNA'] = df['Value'].fillna(df['Value'].mean())

print("Data after filling 'Value' column with mean:")
display(df.head())
```

```python
### Renaming Columns
# Rename a column for better readability or consistency
# Using a dictionary mapping old column names to new column names
# inplace=True modifies the original DataFrame
df.rename(columns={'Date': 'Sale Date'}, inplace=True)

print("Column names after renaming:")
print(df.head())
```

## 4. Data Transformation - Renaming & Type Conversion

In this section, we'll learn how to rename columns and convert data types to match our analysis requirements.

```python
### Type Conversion
# Convert data type of a column to another type
# First fill missing values with the mean, then convert to integer
# This is useful when you need numeric data for calculations
df["Value_new"] = df['Value'].fillna(df['Value'].mean()).astype(int)

print("Data after type conversion to int:")
display(df.head())
```

```python
# Use a lambda function to apply a transformation to each element in the 'Value' column
# In this example, we multiply each value by 2
# Lambda is a quick way to create anonymous functions for simple operations
df["New Value"] = df['Value'].apply(lambda x: x * 2)

print("Data after applying lambda function (multiply by 2):")
display(df.head())
```

### Applying Functions to Columns

The `apply()` method allows you to apply a function to each element in a column. We can use lambda functions for quick, inline operations or custom functions for complex logic.

```python
### Grouping by Single Column
# Group data by 'Product' and calculate the mean of 'Value' for each product
grouped_mean = df.groupby('Product')['Value'].mean()
print("Mean Value by Product:")
display(grouped_mean)

# Group data by 'Product' and calculate the sum of 'Value' for each product
grouped_sum = df.groupby('Product')['Value'].sum()
print("\nSum of Values by Product:")
display(grouped_sum)
```

## 5. Data Aggregation & Grouping

Grouping is a powerful technique to organize data by categories and compute statistics for each group. Common aggregation functions include `sum()`, `mean()`, `count()`, `std()`, and `median()`.

```python
### Grouping by Multiple Columns
# Group by both 'Product' and 'Region' to see sales by product in each region
# This creates a hierarchical grouping with combinations of both columns
grouped_product_region = df.groupby(['Product', 'Region'])['Value'].sum()
print("Sum of Values by Product and Region (Multi-level grouping):")
display(grouped_product_region)
```

```python
### Multiple Aggregation Functions
# Apply multiple aggregate functions to the same grouping
# Using .agg() with a list of function names applies all functions to each group
grouped_aggr = df.groupby("Region")['Value'].agg(['sum', 'mean', 'std', 'median', 'count'])

print("Multiple statistics by Region:")
print("- sum: Total of all values")
print("- mean: Average value")
print("- std: Standard deviation (spread of data)")
print("- median: Middle value")
print("- count: Number of records")
print()
display(grouped_aggr)
```

## 6. Data Joining & Merging

Merging combines two DataFrames based on a common column (key). Different join types control which rows are retained:
- **Inner Join**: Keep only rows where the key exists in both DataFrames
- **Left Join**: Keep all rows from the left DataFrame, fill missing values with NaN
- **Right Join**: Keep all rows from the right DataFrame, fill missing values with NaN
- **Outer Join**: Keep all rows from both DataFrames, fill missing values with NaN

```python
# Create sample DataFrames to demonstrate merging
df1 = pd.DataFrame({
    'Key': ['A', 'B', 'C'],
    'Value1': [1, 2, 3]
})

df2 = pd.DataFrame({
    'Key': ['A', 'B', 'D'],
    'Value2': [4, 5, 6]
})

print("Left DataFrame (df1):")
display(df1)
print("\nRight DataFrame (df2):")
display(df2)

# INNER JOIN: Keep only rows where 'Key' exists in both DataFrames
print("\n--- INNER JOIN ---")
print("Result: Only 'A' and 'B' are kept (they exist in both DataFrames)")
display(pd.merge(df1, df2, on='Key', how='inner'))

# LEFT JOIN: Keep all rows from left DataFrame
print("\n--- LEFT JOIN ---")
print("Result: All rows from df1 (A, B, C), 'D' from df2 is not included")
display(pd.merge(df1, df2, on='Key', how='left'))

# RIGHT JOIN: Keep all rows from right DataFrame
print("\n--- RIGHT JOIN ---")
print("Result: All rows from df2 (A, B, D), 'C' from df1 is not included")
display(pd.merge(df1, df2, on='Key', how='right'))

# OUTER JOIN: Keep all rows from both DataFrames
print("\n--- OUTER JOIN ---")
print("Result: All unique keys from both DataFrames (A, B, C, D)")
display(pd.merge(df1, df2, on='Key', how='outer'))
```
