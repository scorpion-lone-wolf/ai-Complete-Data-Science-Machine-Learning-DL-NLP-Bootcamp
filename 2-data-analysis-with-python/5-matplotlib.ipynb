# Matplotlib

Matplotlib is a Python library for creating static plots, charts, and visualizations.
It is commonly used in data analysis to explore patterns and present results clearly.

```python
import matplotlib.pyplot as plt  # Core Matplotlib plotting module
```

## 1. Line Plots

Line plots are useful for showing how a value changes over an ordered sequence.

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create a basic line plot showing y as a function of x
plt.plot(x, y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Basic Line Plot")
plt.grid(True)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create a styled line plot with markers and dashed lines
plt.plot(
    x,
    y,
    color='red',
    linestyle='--',
    linewidth=2,
    marker='o',
    markerfacecolor='yellow'
)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Customized Line Plot")
plt.grid(True)
plt.show()
```

## 2. Multiple Plots

Use subplots to show more than one chart in the same figure.

```python
# Create multiple plots in a single figure using subplots
x = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]

# figsize controls the width and height of the figure in inches
plt.figure(figsize=(10, 5))

# Create the first subplot in a 1x2 layout
plt.subplot(1, 2, 1)
plt.plot(x, y1, color='green')
plt.title("Linear Relationship")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)

# Create the second subplot in a 1x2 layout
plt.subplot(1, 2, 2)
plt.plot(x, y2, color='blue')
plt.title("Quadratic Relationship")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)

plt.tight_layout()
plt.show()
```

## 3. Bar Plot

Bar charts compare values across categories.

```python
# Bar plot to compare values across categories
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 8, 3, 12]

plt.bar(categories, values, color='skyblue')
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Plot of Category Values')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

## 4. Histogram

Histograms show how data is distributed across value ranges.

```python
# Histogram shows distribution of numerical data by grouping values into bins
ages = [10, 11, 12, 12, 13, 13, 13, 14, 15]

plt.hist(ages, bins=6, color='orange', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution Histogram')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

## 5. Scatter Plot

Scatter plots help visualize the relationship between two numeric variables.

```python
# Scatter plot shows the relationship between two numeric variables
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]

plt.scatter(x, y, marker='x', color='blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Basic Scatter Plot')
plt.grid(True)
plt.show()
```

## 6. Pie Chart

Pie charts display proportions of a whole.

```python
# Pie chart shows the relative proportions of each category
labels = ["Google", "OpenAI", "Anthropic", "Others"]
sizes = [25, 30, 35, 5]
colors = ['blue', 'pink', 'orange', 'green']
explode = [0, 0, 0.1, 0]  # emphasize one slice

plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',
    shadow=True,
    explode=explode
)
plt.title('Market Share Distribution')
plt.show()
```

## 7. Sales Data Example

This section shows how to plot aggregated data from a CSV file.

```python
# Example: Sales data visualization using a CSV file
import pandas as pd

sales_data_df = pd.read_csv('sales_data.csv')
print('Loaded sales data:')
display(sales_data_df.head())
```

```python
# Compute total revenue for each product category
total_sales_by_product = sales_data_df.groupby('Product Category')['Total Revenue'].sum()
print('Total revenue by product category:')
display(total_sales_by_product)
```

```python
# Plot the aggregated revenue as a bar chart
total_sales_by_product.plot(kind='bar', color='teal')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.title('Total Sales by Product Category')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
