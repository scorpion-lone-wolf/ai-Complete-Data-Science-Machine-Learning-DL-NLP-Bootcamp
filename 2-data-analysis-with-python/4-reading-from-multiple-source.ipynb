# Reading Data from Multiple Sources

## Overview
This notebook demonstrates how to read data from various sources using Pandas. In real-world data analysis, data comes from many different formats and locations.

### Topics Covered:
1. **JSON Data** - Reading and writing JSON data
2. **Web Data** - Reading CSV files directly from URLs
3. **HTML Tables** - Extracting tables from web pages
4. **Excel Files** - Reading Excel spreadsheets
5. **Pickle Files** - Serializing and deserializing Python objects

### Prerequisites:
- Basic understanding of Pandas DataFrames
- Familiarity with data formats (JSON, CSV, Excel)
- Internet connection for web-based examples

---

```python
# Import necessary libraries
import pandas as pd  # For data manipulation and analysis
import json          # For JSON data handling
```

## 1. Working with JSON Data

JSON (JavaScript Object Notation) is a popular data format for APIs and web services. Pandas can easily convert JSON data to DataFrames and vice versa.

```python
# Sample JSON data as a string (commonly received from APIs)
json_data = '''
[
  {"name": "Alice", "age": 25, "city": "Delhi"},
  {"name": "Bob", "age": 30, "city": "Mumbai"},
  {"name": "Charlie", "age": 35, "city": "Bangalore"}
]
'''

print("Original JSON data type:", type(json_data))
print("JSON string content:")
print(json_data)

# Convert JSON string to Python objects (list of dictionaries)
data = json.loads(json_data)
print("\nAfter parsing with json.loads():")
print("Type:", type(data))
print("Content:", data)

# Create a Pandas DataFrame from the parsed JSON data
df1 = pd.DataFrame(data)
print("\nConverted to Pandas DataFrame:")
display(df1)
```

```python
# Convert DataFrame back to JSON format
# orient='records' creates a list of dictionaries (most common format)
json_output = df1.to_json(orient='records')

print("DataFrame converted back to JSON:")
print("Type:", type(json_output))
print("Content:", json_output)

# You can also use other orientations:
# orient='split' - separate arrays for index, columns, and data
# orient='index' - dictionary with index as keys
# orient='columns' - dictionary with column names as keys
# orient='values' - just the values as a 2D array
```

## 2. Reading CSV Files from the Web

Pandas can directly read CSV files from URLs without downloading them first. This is useful for accessing public datasets or APIs that return CSV data.

```python
# Read CSV file directly from a URL
# This example uses the UCI Machine Learning Repository's Wine dataset
# header=None means the CSV file doesn't have column headers
# Pandas will automatically assign numeric column names (0, 1, 2, etc.)
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data", header=None)

print("First 5 rows of the Wine dataset:")
display(df.head())

# Save the dataset locally as a CSV file for future use
df.to_csv('wine.csv', index=False)
print("\nDataset saved locally as 'wine.csv'")

# Note: The Wine dataset contains chemical analysis of wines from Italy
# It has 13 chemical attributes and a class label (wine type)
```

## 3. Reading HTML Tables from Web Pages

Many websites contain data in HTML tables. Pandas can extract these tables directly from web pages using `pd.read_html()`. This requires the `requests` library and `lxml` or `beautifulsoup4`.

```python
# Reading tables from HTML web pages
# This requires additional libraries: requests and lxml/beautifulsoup4
import requests
from io import StringIO

# URL of a Wikipedia page with tables
url = "https://en.wikipedia.org/wiki/Mobile_country_code"

# Set user agent to avoid being blocked by websites
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"
}

# Fetch the web page content
response = requests.get(url, headers=headers)

print("Response status code:", response.status_code)
print("Successfully fetched webpage content")

# Extract all tables from the HTML content
# match='Country' filters tables that contain the word "Country"
tables = pd.read_html(StringIO(response.text), match='Country')

print(f"\nFound {len(tables)} table(s) matching 'Country'")
print("First table preview:")
display(tables[0].head())

# Note: pd.read_html() returns a list of DataFrames
# Each DataFrame represents one table found on the page
# You can access specific tables by index: tables[0], tables[1], etc.
```

## 4. Reading Excel Files

Excel files (.xls, .xlsx) are common in business environments. Pandas can read Excel files using `pd.read_excel()`. This requires the `openpyxl` or `xlrd` library.

```python
# Read data from an Excel file
# This assumes you have a file named 'sample.xls' in the current directory
# For .xlsx files, use the same function - Pandas detects the format automatically
df_excel = pd.read_excel("sample.xls")

print("Excel file loaded successfully!")
print("Shape:", df_excel.shape)
print("Columns:", list(df_excel.columns))
print("\nFirst 5 rows:")
display(df_excel.head())

# Additional parameters you can use:
# sheet_name='Sheet1' - specify which sheet to read (default is first sheet)
# header=0 - specify which row contains column headers
# usecols=['A', 'B', 'C'] - specify which columns to read
# skiprows=2 - skip the first 2 rows
```

## 5. Working with Pickle Files

Pickle is Python's way of serializing and deserializing objects. It's useful for saving Python objects (including DataFrames) to disk and loading them back later. Pickle files preserve data types and are faster than CSV for large datasets.

```python
# Save DataFrame to a pickle file
# Pickle preserves all data types and is much faster than CSV for large datasets
# The file extension is typically .pkl or .pickle
df_excel.to_pickle("sample_xls_pickle.pkl")

print("DataFrame saved to pickle file: 'sample_xls_pickle.pkl'")
print("Pickle files are binary and preserve Python object structure")


# Load DataFrame from a pickle file
# This restores the exact DataFrame object that was saved
df_from_pickle = pd.read_pickle('sample_xls_pickle.pkl')

print("DataFrame loaded from pickle file:")
print("Shape:", df_from_pickle.shape)
print("Data types preserved:", df_from_pickle.dtypes.to_dict())
display(df_from_pickle.head())

# Verify that the loaded data matches the original
print("\nData integrity check:")
print("Original and loaded DataFrames are identical:", df_excel.equals(df_from_pickle))
```

## Summary & Key Takeaways

### Data Source Reading Methods:
1. **JSON Data** - Use `json.loads()` then `pd.DataFrame()` for reading, `.to_json()` for writing
2. **Web CSV** - Use `pd.read_csv(url)` to read CSV files directly from URLs
3. **HTML Tables** - Use `pd.read_html()` with `requests` to extract tables from web pages
4. **Excel Files** - Use `pd.read_excel()` to read .xls and .xlsx files
5. **Pickle Files** - Use `.to_pickle()` and `pd.read_pickle()` for fast Python object serialization

### Best Practices:
- **JSON**: Use `orient='records'` for API-compatible format
- **Web Data**: Always set User-Agent headers when scraping websites
- **Excel**: Specify sheet names when working with multi-sheet workbooks
- **Pickle**: Only unpickle files from trusted sources (security risk)
- **Error Handling**: Always check if files exist and handle network errors

### Common File Extensions:
- `.json` - JSON data files
- `.csv` - Comma-separated values
- `.xls/.xlsx` - Microsoft Excel files
- `.pkl/.pickle` - Python pickle files
- `.html` - Web pages with tables

### Performance Tips:
- Pickle files are fastest for Python objects
- CSV is most portable across different systems
- Excel files preserve formatting but are slower to read
- JSON is human-readable and works well with APIs
