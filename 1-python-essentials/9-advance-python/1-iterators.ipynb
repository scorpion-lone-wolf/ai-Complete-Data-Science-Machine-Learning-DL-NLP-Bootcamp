### Iterators 
- Allows for efficient looping and memory management.
- Iterators provides a way to access elements of a collection sequentially without exposing the underlying structure.

```python
my_list=[1,2,3,]
my_iterator=iter(my_list)
print(type(my_iterator))
```

```python
# Way to access elements using next() function , without exposing the underlying structure of the collection
try:
    print(next(my_iterator)) # Output: 1
    print(next(my_iterator)) # Output: 2
    print(next(my_iterator)) # Output: 3
    print(next(my_iterator)) # Raises StopIteration exception as there are no more elements to iterate
except StopIteration:
    print("No more elements to iterate.")
```

```python
# dictionary
my_dict={'a':1,'b':2,'c':3}
dict_iterator=iter(my_dict)
print(type(dict_iterator))
print(next(dict_iterator)) # Output: 'a' (the first key in the dictionary)
print(next(dict_iterator)) # Output: 'b' (the second key in the dictionary)
print(next(dict_iterator)) # Output: 'c' (the third key in the dictionary)
```
