### Decorators 
- It allows you to modify the behaviour of a function or class methods.
- They are commonly used to add functionality to functions or methods without modifying their actual code.

```python
## Function copy
#  It is nothing but treating a function as an object and assigning it to another variable.
# This is possible because in Python, functions are first-class objects, meaning they can be passed around and assigned to variables just like any other object (like integers, strings, etc.).
# When you assign a function to another variable, you are essentially creating a new reference to the same function object in memory.
# This allows you to call the function using either variable name, and both will execute the same code.
def greet(name):
    return f"Hello, {name}!"
greet_copy = greet
print(greet_copy("Alice"))  # Output: Hello, Alice!
```

```python
## closures
# Meaning the inner function has access to the variables of the outer function, even after the outer function has finished executing.
def outer_function(msg):
    message = msg
    def inner_function():
        print(message)
    return inner_function
hi_func= outer_function("Hi") # This will return the inner function, which has access to the message variable
hi_func()  # Output: Hi
```

```python
### Decorators
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper
```

```python
# my normal function
def say_hi():
    print("Hello, world!")

# using decorator
my_decorated_fn = my_decorator(say_hi)
my_decorated_fn()
```

```python
# using decorator using @syntax
@my_decorator
def say_hi():
    print("Hello, world!")

say_hi()
```

### Conclusion
- What really happen is the decorator function take the function as argument and do something with it.
