# Variables in Python

## Declaring and Assigning Variables

In Python, a variable is created the moment you first assign a value to it.

```python
age = 28
height = 5.4
name = 'rahul'
is_student = False

print("Name is:", name)
print("Age is:", age)
print("Height is:", height)
print("Is Student:", is_student)
```

## Variable Naming Conventions

- Variables should be descriptive.
- They must start with a letter or an underscore `_` and can contain letters, numbers, and underscores.
- Variable names are case-sensitive.

### Valid variable names:

```python
first_name = 'Rahul'
_last_name = 'Singh'
age1 = 28
```

### Invalid variable names:

```python
# 2name = 'rahul'    # Cannot start with a number
# first-name = 'rahul' # Cannot contain hyphens
# @name = 'rahul'      # Cannot contain special characters like '@'
```

## Variable Types

Python is a dynamically typed language, which means the type of a variable is determined at runtime.

```python
age = 28         # int
height = 5.34    # float
name = 'rahul'    # str
is_student = True # bool
```

### Type Checking and Conversion

You can check the type of a variable using the `type()` function and convert between types using functions like `int()`, `float()`, `str()`, etc.

```python
# Type checking
print(type(name))
print(type(age))
print(type(height))
```

```python
# Type Conversion
print(int(1.23))      # float to int
print(int(False))     # bool to int
print(int('23'))      # str to int
print(float(1))       # int to float
```

## Dynamic Typing

Python allows the type of a variable to change as the program executes.

```python
var = 10
print(var, type(var))

var = 'rahul'
print(var, type(var))
```

## Example: Simple Calculator

```python
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Perform calculations
sum_val = num1 + num2
diff_val = num1 - num2
mul_val = num1 * num2
exp_val = num1 ** num2

print("Sum:", sum_val)
print("Difference:", diff_val)
print("Multiplication:", mul_val)
print("Exponentiation:", exp_val)
```
