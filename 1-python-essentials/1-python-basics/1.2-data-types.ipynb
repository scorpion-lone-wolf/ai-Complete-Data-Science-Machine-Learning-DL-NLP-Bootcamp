# Data Types in Python

In programming, data type is an important concept. Variables can store data of different types, and different types can do different things.

## Integer (`int`)

Integers are whole numbers, positive or negative, without decimals.

```python
age = 20
print(type(age))
```

## Float (`float`)

Floats are numbers, positive or negative, containing one or more decimals.

```python
height = 5.4
print(type(height))
```

## String (`str`)

Strings are sequences of characters, enclosed in single or double quotes.

```python
name = 'rahul'
print(type(name))
```

## Boolean (`bool`)

Booleans represent one of two values: `True` or `False`.

```python
is_male = True
print(type(is_male))
```

## Type Casting and String Concatenation

You can't concatenate a string with a number directly. You need to convert the number to a string first using `str()`.

```python
# Incorrect way (will cause a TypeError)
# print("hello" + 5)

# Correct way
print("hello" + str(5))
```
