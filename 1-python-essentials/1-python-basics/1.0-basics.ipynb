# Python Basics

## Indentation in Python

Indentation in Python is used to define the structure and hierarchy of the code.

Unlike many other programming languages that use braces `{}` to delimit blocks of code, Python uses indentation to determine the grouping of statements.

This means that all the statements within a block must be indented at the same level (usually 4 spaces).

```python
age = 32
if age > 30:
    print(age)
print(age)
```

### Code Example of indentation

```python
if True:
    print("This is Example of indentation")
    if False:
        print("This will not print")
    print("This will be printed")
print("This is outside of if block")
```

## Line Continuation in Python

Python uses a backslash (`\`) to continue a statement to the next line.

```python
total = 1 + 2 + 3 + 4 + \
        5 + 6
print("Total is ", total)
```

## Variable Assignment
In Python, you don't need to declare variables before using them. A variable is created the moment you first assign a value to it.

```python
name = 'rahul'  # name is a string
age = 28      # age is an integer
print(f'Type of name is {type(name)} and Type of age is {type(age)}')
```
