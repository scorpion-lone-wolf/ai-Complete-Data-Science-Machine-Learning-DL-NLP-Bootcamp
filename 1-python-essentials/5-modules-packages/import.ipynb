# Python Modules and Packages

Short notes on different ways to import in Python.

- `import module` imports the whole module.
- `from module import name` imports only the required item.
- `import module as alias` gives the module a shorter name.
- Packages help us organize related modules inside folders.

## 1. Import the full module

Use this when you want to keep it clear where a function comes from.

```python
import math

num1 = 23
num2 = 33

print("Square root:", math.sqrt(num1))
print("Power:", math.pow(num1, num2))
```

## 2. Import only what you need

This keeps the code short when you only need one or two names.

```python
from math import pi

print("Value of pi:", pi)
```

## 3. Import with an alias

Aliases are common for libraries with long names. `numpy as np` is the standard style.

```python
import numpy as np

numbers = np.array([1, 2, 3])
print(numbers)
```

## 4. Wildcard import

This works, but it is usually **not recommended** because it imports many names into the current namespace and can make the code harder to read.

```python
from math import *

print("sqrt(25) =", sqrt(25))
```

## 5. Import from our own package

Here, `package` is a folder and `calc.py` is a module inside it.
The file `__init__.py` makes the folder behave like a regular Python package.

```python
from package.calc import add, subtract

print("Addition:", add(2, 3))
print("Subtraction:", subtract(2, 3))
```

## Quick summary

- Use `import module` for clarity.
- Use `from module import name` when you need only a few items.
- Use aliases like `np` for common libraries.
- Avoid `from module import *` in real projects when possible.
- Use packages to group related modules.
