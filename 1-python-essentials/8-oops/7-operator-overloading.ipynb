### Operator Overloading
- Operator Overloading allows you to define the behaviour of operators ( + , - , * , etc) for custom objects.
- You achieve this by overriding specific magic methods in your class

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        return f"Hello, my name is {self.name} and I am {self.age} years old."
    # operator overloading
    def __add__(self, other):
        if isinstance(other, Person):
            return f"{self.name} and {other.name} are friends."
        return NotImplemented
    def __eq__(self,other):
        if isinstance(other,Person):
            return "Both are of same age" if self.age == other.age else "They are of different ages"
        return NotImplemented
```

```python
#  Example Usage
person1=Person("Alice", 30)
person2=Person("Bob", 30)
print(person1.greet())
# addition of two person objects
print(person1 + person2) # this calls the __add__ method
# equality check of two person objects
print(person1 == person2) # this calls the __eq__ method
```
