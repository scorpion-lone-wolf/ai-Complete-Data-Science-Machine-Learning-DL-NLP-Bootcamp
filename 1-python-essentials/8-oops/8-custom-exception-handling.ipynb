#### Custom Exception
- In this we create our own custom exception which extend base Exception class
- we raise the custom exception

```python
class UnderAgeError(Exception):
    pass
```

```python
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def can_vote(self):
        if self.age < 18:
            raise UnderAgeError("You must be at least 18 years old to vote.")
        return "You are allowed to vote."
```

```python
person1=Person("Alice", 17)
try:
    print(person1.can_vote())
except UnderAgeError as e:
    print(e)
```

```python
person2=Person("Bob", 20)
try:
    print(person2.can_vote())
except UnderAgeError as e:
    print(e)
```
