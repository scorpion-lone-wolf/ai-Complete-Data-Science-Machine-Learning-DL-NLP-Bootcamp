### Magic Methods
- Magic methods are predefined methods in python that you can override to change the behaviour of your object.

```python
# Examples :
# __init__ : Initialize a new instance of a class.
# __str__ : Return a string representation of an object.
# __len__ : Return the length of an object.
# __add__ : Define the behavior of the addition operator (+) for a class.
# __eq__ : Define the behavior of the equality operator (==) for a class.
# __getitem__ : Define the behavior of indexing (e.g., obj[key]) for a class.
# __setitem__ : Define the behavior of item assignment (e.g., obj[key] = value) for a class.
# __iter__ : Define the behavior of an object when it is iterated over (e.g., in a for loop).
# __next__ : Define the behavior of an iterator when the next() function is called on it.
# __repr__ : Return a string representation of an object that is meant to be unambiguous and can be used to recreate the object.
# __del__ : Define the behavior of an object when it is about to be destroyed (called when an object is garbage collected).
# __contains__ : Define the behavior of the 'in' operator for a class.
# __bool__ : Define the behavior of an object when it is evaluated in a boolean context (e.g., in an if statement).
# __hash__ : Define the behavior of an object when it is used as a key in a dictionary or stored in a set.
```

```python
class MyClass:
    def __init__(self, value):
        self.value = value

    def __str__(self):
        return f"MyClass with value: {self.value}"

    def __len__(self):
        return len(str(self.value))

    def __add__(self, other):
        if isinstance(other, MyClass):
            return MyClass(self.value + other.value)
        return NotImplemented

    def __eq__(self, other):
        if isinstance(other, MyClass):
            return self.value == other.value
        return NotImplemented

    def __getitem__(self, key):
        return str(self.value)[key]

    def __setitem__(self, key, value):
        value_str = str(self.value)
        if key < len(value_str):
            self.value = value_str[:key] + str(value) + value_str[key + 1:]

    def __iter__(self):
        return iter(str(self.value))

    def __next__(self):
        print("This is the next method, but it is not implemented as an iterator.")
        # raise StopIteration

    def __del__(self):
        print(f"MyClass with value {self.value} is being destroyed")

    def __contains__(self, item):
        return str(item) in str(self.value)

    def __bool__(self):
        return bool(self.value)

    def __hash__(self):
        return hash(self.value)
```

```python
# create an instance of MyClass and demonstrate the use of the special methods
my_class = MyClass(12345) #  = this will call the __init__ method to initialize the instance with the value 12345
print(len(my_class))  # 5 = this will call the __len__ method to return the length of the string representation of the value (which is 5)
print(bool(my_class))  # True = this will call the __bool__ method to evaluate the instance in a boolean context (which is True because the value is not empty or zero)
print(5 in my_class)  # True = this will call the __contains__ method to check if the string '5' is in the string representation of the value (which is True)
print(my_class + MyClass(67890))  # MyClass with value: 80235 = this will call the __add__ method to add the values of two MyClass instances and return a new MyClass instance with the result (which is 80235)
print(hash(my_class))  # this will call the __hash__ method to return the hash value of the instance (which is based on the value 12345)
next(my_class)
print(my_class)  # MyClass with value: 12345 = this will call the __str__ method to return a string representation of the instance (which is "MyClass with value: 12345")
# del my_class  # this will call the __del__ method to print a message when the instance is being destroyed

```
