# Map function notes

- `map()` applies a callable to each item of an iterable and returns an iterator.
- Convert the returned iterator when you need repeated access or a concrete sequence (e.g., `list()`).

## Applying a named function to every item

Define a helper function and then pass it to `map()` for clarity when the transformation is more than one line.

```python
lst = [1, 2, 3, 4, 5, 6]

def square(x):
    return x ** 2

# map applies `square` lazily across the iterable
squared_lst = map(square, lst)
print(list(squared_lst))
```

## Inline transformation with lambda

Use a lambda when the logic fits in one expression to keep the definition next to the call.

```python
# same behavior but defined inline
squared_lst_lambda = map(lambda x: x ** 2, lst)
print(list(squared_lst_lambda))
```

## Combining multiple iterables

`map()` accepts multiple iterables. The callable must accept matching arguments, and iteration stops at the shortest iterable.

```python
lst1 = [1, 2, 3]
lst2 = [4, 5, 6]

mul_list = map(lambda x, y: x * y, lst1, lst2)
print(list(mul_list))
```

## Type conversion with built-ins

Passing built-in constructors (e.g., `int`) to `map()` quickly sanitizes string data.

```python
str_number = ["1", "2", "3", "4", "5"]

mapped_numbers = map(int, str_number)
print("before:", str_number)
print("after:", list(mapped_numbers))
```

## Applying transformations to structured data

Pull specific fields out of dictionaries or objects using `map()` and a lambda.

```python
peoples = [
    {"name": "Alice", "age": 30},
    {"name": "Bob", "age": 25},
    {"name": "Charlie", "age": 35}
]

names = list(map(lambda person: person['name'], peoples))
ages = list(map(lambda person: person['age'], peoples))

print("names:", names)
print("ages:", ages)
```

## String manipulation example

Any sequence transformation works; here we uppercase every string before printing.

```python
str_list = ["hello", "world", "python"]
upper_str_list = list(map(lambda s: s.upper(), str_list))
print(upper_str_list)
```

## Key takeaways

- `map()` is lazy; wrap with `list()`, `tuple()`, etc., when you need a reusable container.
- The callable and each iterable must agree on the number of arguments.
- Combine `map()` with constructors or attribute access to clean or reshape data succinctly.
