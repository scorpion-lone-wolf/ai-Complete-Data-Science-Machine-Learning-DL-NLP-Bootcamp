# Filter function notes

- `filter()` keeps only the items for which the supplied predicate returns `True` and returns an iterator.
- Combine `filter()` with `map()` or comprehensions when you need to transform filtered results.

## Chaining `filter()` and `map()`

First filter even numbers, then square each remaining item with `map()`. Wrapping the whole pipeline in `list()` materializes the iterator for printing.

```python
lst = list(range(1, 11))  # 1 to 10 inclusive

# keep even numbers first, then square them
filtered = filter(lambda x: x % 2 == 0, lst)
squared_filter_list = list(map(lambda x: x ** 2, filtered))
print(squared_filter_list)
```

## Filtering with multiple conditions

`filter()` can combine predicates: under the hood you can use `and`, `or`, and additional checks to capture the exact subset you want.

```python
lst = list(range(1, 11))

# keep numbers that are even and greater than five
filtered_list = list(filter(lambda x: x % 2 == 0 and x > 5, lst))
print(filtered_list)
```

## Key takeaways

- `filter()` is lazy so wrap it with `list()`/`tuple()`/`set()` when you need the results multiple times.
- You can pass any callable that returns a boolean; lambdas keep inline logic short.
- Combine `filter()` and `map()` to both select and transform data in a readable pipeline.
