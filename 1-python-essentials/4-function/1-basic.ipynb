# Functions Basics

This notebook introduces Python functions with small, clear examples.


## Function Syntax
A function groups reusable logic. The **docstring** describes what it does.


```python
def function_name(parameters):
    """Docstring: describe the function here."""
    # Function body
    return value
```

## Why Use Functions?
- Reusability: write once, use many times.
- Modularity: break big problems into smaller pieces.
- Readability: meaningful names improve clarity.
- Abstraction: hide details behind a simple interface.


## Defining and Calling a Function


```python
def add(a: int, b: int) -> int:
    """Return the sum of two numbers."""
    return a + b

# Call the function
print(add(2, 3))
print(add(5, 7))
```

## Parameters vs Arguments
- **Parameters** are variables in the function definition.
- **Arguments** are the actual values passed when calling the function.


```python
def square_of_number(num: int) -> int:
    """Return the square of a number."""
    return num ** 2

# Here `num` is a parameter and 5 is an argument.
print(square_of_number(5))
```

## Default Parameters
If the caller does not provide a value, the default is used.


```python
def greet(name: str = 'Rahul', msg: str = 'Hello') -> str:
    """Greet a person with a message."""
    return f'{msg} {name}'

print(greet())
print(greet('Alice'))
print(greet('Bob', 'Hi'))
```

## Named (Keyword) Arguments
Keyword arguments improve readability and allow any order.


```python
def greet(name: str = 'Rahul', msg: str = 'Hello') -> str:
    """Greet a person with a message."""
    return f'{msg} {name}'

print(greet(msg='Hi', name='Alice'))
```

## Positional Arguments and *args
`*args` collects extra positional arguments into a tuple.


```python
numbers = [1, 2, 3, 4, 5]

def sum_number(num1: int, *num2: int) -> int:
    """Return the sum of all numbers."""
    total = num1 + sum(num2)
    return total

print(sum_number(*numbers))
```

## **kwargs (Keyword Arguments Dictionary)
`**kwargs` collects named arguments into a dictionary.


```python
def info(**kwargs):
    print(kwargs)

info(name='Alice', age=30, city='New York')
```

## Mixing *args and **kwargs
You can accept both positional and keyword arguments together.


```python
def info(*args, **kwargs):
    print(args)
    print(kwargs)

info(1, 2, 3, name='Alice', age=30, city='New York')
```
