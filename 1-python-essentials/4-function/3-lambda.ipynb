# Python Lambda Functions: A Comprehensive Guide

This notebook explores `lambda` functions in Python. We'll cover their syntax, use cases, and how they integrate with other built-in functions like `map` and `filter`.

## 1. What are Lambda Functions?

A lambda function is a small, **anonymous function** defined with the `lambda` keyword. They are also known as lambda expressions or lambda abstractions.

Key characteristics:
- **Anonymous:** They don't have a name like functions defined with `def`.
- **Single Expression:** They can have any number of arguments but can only contain a single expression. This expression is evaluated and returned when the function is called.
- **Concise:** They are ideal for simple, one-off operations where defining a full function with `def` would be unnecessarily verbose.

### Syntax

The basic syntax of a lambda function is:

```python
lambda arguments: expression
```

````python
# A regular function
def square_def(x):
    return x ** 2

# The equivalent lambda function
# We assign it to a variable to be able to call it.
square_lambda = lambda x: x ** 2

print(f"Using def: {square_def(5)}")
print(f"Using lambda: {square_lambda(5)}")
````

## 2. Core Concepts & Examples

### With Multiple Arguments

````python
# This lambda function takes two arguments, x and y
multiply = lambda x, y: x * y

print(f"5 * 4 = {multiply(5, 4)}")
````

### With Conditional Logic

Lambda functions can use a ternary (conditional) expression to perform simple `if-else` logic.

````python
# Checks if a number is even or odd
even_or_odd = lambda x: 'Even' if x % 2 == 0 else 'Odd'

print(f"5 is {even_or_odd(5)}")
print(f"6 is {even_or_odd(6)}")
````

### Lambda Functions are Still Functions

Even though they are defined differently, a lambda expression creates a standard function object.

````python
print(type(even_or_odd))
````

## 3. Usage with Higher-Order Functions

The primary use case for lambda functions is as an argument to **higher-order functions**, which are functions that take other functions as arguments (e.g., `map`, `filter`, `sorted`).

### `map()`

The `map()` function applies a given function to every item of an iterable (like a list) and returns a map object (an iterator).

**Syntax:** `map(function, iterable)`

````python
numbers = [1, 2, 3, 4, 5]

# Use a lambda to square each number in the list
squared_numbers_iterator = map(lambda x: x ** 2, numbers)

print(f"Map object: {squared_numbers_iterator}")

# To see the result, we convert the iterator to a list
squared_numbers_list = list(squared_numbers_iterator)
print(f"Squared numbers: {squared_numbers_list}")
````

### `filter()`

The `filter()` function creates an iterator from elements of an iterable for which a function returns `True`.

**Syntax:** `filter(function, iterable)`

````python
numbers = [1, 2, 3, 4, 5, 6, 7, 8]

# Use a lambda to filter out only the even numbers
even_numbers_iterator = filter(lambda x: x % 2 == 0, numbers)

print(f"Filter object: {even_numbers_iterator}")

# Convert the iterator to a list to view the results
even_numbers_list = list(even_numbers_iterator)
print(f"Even numbers: {even_numbers_list}")
````

### Sorting with `sorted()`

Lambda functions are extremely useful for providing a custom sorting `key` to the `sorted()` function or the `.sort()` list method. This is especially powerful when sorting complex objects like lists of tuples or dictionaries.

````python
# A list of tuples (student_name, age, score)
students = [('Alice', 22, 88), ('Bob', 25, 95), ('Charlie', 20, 82)]

# Sort by age (the second item in each tuple, index 1)
sorted_by_age = sorted(students, key=lambda student: student[1])
print(f"Sorted by age: {sorted_by_age}")

# Sort by score (the third item, index 2) in descending order
sorted_by_score_desc = sorted(students, key=lambda student: student[2], reverse=True)
print(f"Sorted by score (desc): {sorted_by_score_desc}")
````

## 4. Conclusion: When to Use Lambda Functions

Lambda functions are a powerful and concise tool, but they are not always the right choice.

**Use a `lambda` when:**
- You need a small, temporary function to pass to a higher-order function.
- The logic is simple and can be expressed in a single line.
- Readability is improved by keeping the function definition right where it's used (e.g., inside a `sorted()` call).

**Use `def` (a named function) when:**
- The logic is too complex for a single expression.
- You need to reuse the function in multiple places.
- The function requires a docstring, type hints, or a descriptive name to be understood.
- You are writing code that needs to be easily debugged (stack traces from named functions are more informative than those from lambdas).
