# Tuples in Python

## Introduction to Tuples
Tuples are ordered, immutable collections of items. They are defined using parentheses `()` and can contain elements of different data types. Tuples are similar to lists but cannot be modified after creation, making them useful for storing fixed collections of data.

## Creating Tuples

```python
my_tuple = (1, 2, 3, 'a', 'b', 'c')
print(my_tuple)
print(type(my_tuple))
```

## Accessing Tuple Elements

```python
print(my_tuple[0])  # Output: 1
print(my_tuple[3])  # Output: 'a'
print(my_tuple[-1]) # Output: 'c'
print(my_tuple[1:4])  # Output: (2, [3, 4], 'a')
print(my_tuple[::-1]) # reverse the tuple # Output: ('c', 'b', 'a', [3, 4], 2, 1)
```

## Tuple Operation

```python
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
# Concatenation
tuple3 = tuple1 + tuple2
print(tuple3)  # Output: (1, 2, 3, 4, 5, 6)
# Repetition
tuple4 = tuple1 * 2
print(tuple4)  # Output: (1, 2, 3, 1, 2, 3)
```

## Immutable Nature of Tuples

```python
lst = [1, 2, 3]
tup = (1, 2, 3)
lst[0] = 10  # This works, lists are mutable
print(lst)  # Output: [10, 2, 3]
try:
    tup[0] = 10  # This will raise an error, tuples are immutable
except TypeError as e:
    print(e)  # Output: 'tuple' object does not support item assignment
```

## Tuple methods

```python
my_tuple = (1, 2, 3, 'a', 'b', 'c', 2)
print(my_tuple.count(2))  # Output: 2
print(my_tuple.index('a'))  # Output: 3
```

## Packing and Unpacking tuples

```python
my_tuple= 1, 2, 3, 'a', 'b', 'c'  # Packing
print(my_tuple, type(my_tuple))  # Output: (1, 2, 3, 'a', 'b', 'c') <class 'tuple'>

first,second,*rest= my_tuple  # Unpacking
print(first)  # Output: 1
print(second) # Output: 2
print(rest, type(rest))   # Output: [3, 'a', 'b', 'c'] <class 'list'>
```

## Nested Tuples

```python
nested_tuple = (
    (1, 2, 3),
    ('a', 'b', 'c'),
    (4.5, 5.5, 6.5),
    ("rahul", "sachin", "virat")
)
# Accessing elements in nested tuples
print(nested_tuple[0][1])  # Output: 2
print(nested_tuple[1][2])  # Output: 'c'
print(nested_tuple[2][0])  # Output: 4.5
print(nested_tuple[3][1])  # Output: 'sachin'
```

```python
## Membership Testing
my_tuple = (1, 2, 3, 'a', 'b', 'c')
print(2 in my_tuple)  # Output: True
print('d' in my_tuple)  # Output: False
```

```python
## Tuples immuatible but can be modified if list is inside tuple
my_tuple = (1, 2, [3, 4], 'a', 'b', 'c')
print(my_tuple)  # Output: (1, 2, [3, 4], 'a', 'b', 'c')
my_tuple[2].append(10)  # Modifying the list inside the tuple
print(my_tuple)
```

## Practical Examples of Tuples

```python
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=' ')
    print()  # for new line after each sub-tuple
```
