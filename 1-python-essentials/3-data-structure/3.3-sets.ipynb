# Sets in Python

Sets are built-in data structures in Python that are unordered collections of unique elements. They are useful for storing and manipulating groups of items without duplicates. Sets are defined using curly braces `{}` or the built-in `set()` function.

Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union, intersection, difference, and symmetric difference.

## Creating Sets

```python
my_set = {1, 2, 3, 4, 5}
print(my_set)
print(type(my_set))
```

### Creating an Empty Set

```python
empty_set = set()
print(empty_set)
```

### Creating a Set from a List

```python
my_list = [1, 2, 3, 4, 5, 5, 4]
my_set_from_list = set(my_list)
print(my_set_from_list)
```

## Set Properties
### Unique Elements
Sets automatically handle duplicates.

```python
my_set = {1, 2, 3, 4, 5}
my_set.add(3) # Trying to add a duplicate element
my_set.add(5) # Trying to add another duplicate element
print(my_set)
```

## Modifying Sets

### Adding and Removing Elements

```python
my_set = {1, 2, 3, 4, 5}
print("Original set:", my_set)

# Adding an element
my_set.add(6)
print("After adding 6:", my_set)

# Removing an element
my_set.remove(2)
print("After removing 2:", my_set)

# `discard()` will not raise an error if the element is not found
my_set.discard(10)
print("After trying to discard 10:", my_set)
```

### Clearing a Set

```python
print("Set before clearing:", my_set)
my_set.clear()
print("Set after clearing:", my_set)
```

## Set Operations

### Membership Test

```python
my_set = {1, 2, 3, 4, 5}
print("Is 3 in my_set?", 3 in my_set)
print("Is 6 in my_set?", 6 in my_set)
```

## Mathematical Set Operations

```python
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}
```

### Union

The union of two sets (`set_a.union(set_b)` or `set_a | set_b`) results in a new set containing all the unique elements from **both** sets.

```python
union_ab = set_a.union(set_b)
print("Union of set_a and set_b:", union_ab)
```

### Intersection

The intersection of two sets (`set_a.intersection(set_b)` or `set_a & set_b`) results in a new set containing only the elements that are common to **both** sets.

```python
intersection_ab = set_a.intersection(set_b)
print("Intersection of set_a and set_b:", intersection_ab)
```

### Difference

The difference between two sets (`set_a.difference(set_b)` or `set_a - set_b`) results in a new set containing elements that are in `set_a` but **not** in `set_b`. 

**Note:** The order matters. `set_a.difference(set_b)` is not the same as `set_b.difference(set_a)`.

```python
difference_ab = set_a.difference(set_b)
print("Difference of set_a and set_b (a - b):", difference_ab)
difference_ba = set_b.difference(set_a)
print("Difference of set_b and set_a (b - a):", difference_ba)
```

### Symmetric Difference

The symmetric difference of two sets (`set_a.symmetric_difference(set_b)` or `set_a ^ set_b`) results in a new set containing elements that are in **either** `set_a` **or** `set_b`, but **not** in both. 

Think of it as the elements that are unique to each set.

```python
symmetric_difference_ab = set_a.symmetric_difference(set_b)
print("Symmetric difference:", symmetric_difference_ab)
```

## Other Set Methods

```python
set1 = {1, 2, 3, 4, 5}
set2 = {3, 4, 5}
print("set2 is a subset of set1:", set2.issubset(set1))
print("set1 is a superset of set2:", set1.issuperset(set2))
```

## Practical Example
### Counting Unique Words in Text

```python
text = "In this tutorial we are discussing about sets"
print("Original text:", text)
unique_words = set(text.split())
print("Unique words:", unique_words)
print("Number of unique words:", len(unique_words))
```
