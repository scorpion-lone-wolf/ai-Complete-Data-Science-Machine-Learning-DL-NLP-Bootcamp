# File Operations in Python

This notebook covers fundamental file operations in Python, including reading from files, writing to files, and copying file contents.

---

```python
# ============================================================
# 1. READING FROM A FILE
# ============================================================

# Before running:
# Make sure you have a file named 'example.txt' in the same directory
# with content like:
# This is line 1
# This is line 2
# This is line 3
# This is line 4
# This is line 5


# ------------------------------------------------------------
# 1.1 Reading the Entire File at Once
# ------------------------------------------------------------

# NOTE:
# - This reads the entire file into memory as a single string.
# - Not recommended for very large files.

with open('example.txt', 'r') as file:
    # Read the entire content
    content = file.read()

    # Print content
    print("Full File Content:\n")
    print(content)


# ------------------------------------------------------------
# 1.2 Reading File Line by Line (Memory Efficient)
# ------------------------------------------------------------

# NOTE:
# - This is better for large files
# - Processes one line at a time

with open('example.txt', 'r') as file:
    print("\nReading line by line:\n")

    for line in file:
        # strip() removes extra spaces and newline characters
        print(line.strip())

```

```python

# ============================================================
# 2. WRITING TO A FILE
# ============================================================

# ------------------------------------------------------------
# 2.1 Creating and Writing to a File
# ------------------------------------------------------------

# MODE EXPLANATION:
# 'w' → Write mode (creates new file OR overwrites existing file)
# 'a' → Append mode (adds content without deleting existing content)

with open('content.txt', 'w') as file:
    # Writing multiple lines to file
    file.write("This is the first line.\n")
    file.write("This is the second line.\n")
    file.write("This is the third line.\n")
    file.write("This is the fourth line.\n")

print("\nFile 'content.txt' has been created/overwritten.")
```

```python

# ============================================================
# 3. COPYING FILE CONTENT
# ============================================================

# This copies content from 'example.txt' → 'copy.txt'

with open('example.txt', 'r') as source_file, open('copy.txt', 'w') as dest_file:

    for line in source_file:
        dest_file.write(line)

print("\nFile copied successfully to 'copy.txt'.")
```

```python
lines = [
    "Line 1\n",
    "Line 2\n",
    "Line 3\n"
]

with open('content.txt', 'w') as file:
    file.writelines(lines)
```
