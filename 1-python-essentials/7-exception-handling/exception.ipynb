## Exception Handling in Python

Exception handling allows you to handle runtime problems gracefully and take corrective actions without abruptly stopping the program.

### Exception

An exception is an event that disrupts the normal flow of a program during execution.
It usually occurs when a runtime error happens.

### Error

An error is a broader term for serious problems in a program.
Some errors can be handled as exceptions, while others cannot.

All exceptions are errors/problems, but not all errors are exceptions.

```python
# Assign values to two numbers
num1 = 5
num2 = 0

# Try to divide num1 by num2
try:
    result = num1 / num2
    print("The result is:", result)

# Handle division by zero error
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")

# Handle any other unexpected errors
except Exception as e:
    print("An error occurred:", e)
```

```python
# Demonstration of try, except, else, and finally blocks

# Assign values to two numbers
num1 = 50
num2 = 2

try:
    # Perform division
    result = num1 / num2

    # Raise a custom exception manually
    raise Exception("This is a custom exception.")

# Handle division by zero error
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")

# Handle any other exception
except Exception as e:
    print("An error occurred:", e)

# Executes only if no exception occurs in try block
else:
    print("The result is:", result)

# Executes always, whether exception occurs or not
finally:
    print("This block will always execute.")
```
